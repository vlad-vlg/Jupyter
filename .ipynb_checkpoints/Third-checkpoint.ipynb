## Покажем на примере различие в скорости работы массивов и списков.
 Посчитаем сумму квадратных корней первых $10^7$ чисел:  
 <h3><center> $y=\sum_{i=0}^{10^7}  \sqrt x$ </center></h3>

```python
import numpy as np
from time import time

t = time()
print(f"Результат итератора: {sum(x ** 0.5 for x in range(10 ** 7))}.")
print(f"Время выполнения: {time() - t} с.")
print()
t = time()
print(f"Результат numpy: {np.sqrt(np.arange(10 ** 7)).sum()}.")
print(f"Время выполнения: {time() - t} с.")
```

Библиотека *`numpy`* решила задачу в 20 раз быстрее, чем итератор.

# Обработка .csv файла при помощи pandas

### Загружаем файл данных

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/10_Invoices.csv', sep=';')
df
```

### Приводим колонку "***Счет***" к однообразному виду и переименуем её в "***Наименование курса***"

```python
cnt = df[df.columns[0]].count()
for i in range(cnt):
    df.iat[i, 1] = df.iat[i, 1].replace('Сделка по заявке ', '').replace('Запись на к', 'К')
df.rename(columns={'Счет': 'Наименование курса'}, inplace=True)
df.tail(10)
```

### Посмотрим статистику по количеству записей на курсы:

```python
courses = df.groupby('Наименование курса').size().sort_values(ascending=True).reset_index(name='Количество')
courses
```

```python
fig, ax1 = plt.subplots( figsize=(7,3))
ax1.barh(y=courses['Наименование курса'], color="m", width=courses['Количество'])

for i, val in enumerate(courses['Количество'].values):
    plt.text(val, i, val, horizontalalignment='left', verticalalignment='center', fontdict={'size':10, 'color':'r'})

ax1.set_title('Количество записавшихся на курсы', fontdict={'fontsize': 14, 'fontweight': 20, 'color':'b'})
ax1.set_yticks(courses['Наименование курса'])
ax1.set(xlim=(0, 70))
plt.ylabel('Наименование курса', color='gray')
plt.xlabel('Количество',color='gray');
```

### Посмотрим статус оплаты обучения:

```python
status = df.groupby('Статус').size().sort_values(ascending=True).reset_index(name='Количество')
print(status)
colors = ('m', 'b', 'y', 'g')
fig, ax2 = plt.subplots( figsize=(4,4))
ax2.set_title('Статус оплаты обучения', fontdict={'fontsize': 14, 'fontweight': 20, 'color':'b'})
ax2.pie(status['Количество'],
        explode=[0,0,0,.03],
        wedgeprops={"linewidth": 1, "edgecolor": "white"},
        labels=status['Статус'],
        colors=colors,
        startangle=70,
        autopct = '%1.2f%%');
```
